Khai báo thư viện cần thiết

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

Load ảnh

```python
img = mpimg.imread('girl3.jpg')
print type(img)
print img.shape
plt.imshow(img)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show() 
```

Reshape bức ảnh thành ma trận một chiều, mỗi phần tử là 1 pixel với 3 giá trị màu

```python
X = img.reshape((img.shape[0] * img.shape[1], img.shape[2]))
print X.shape
```

Segmentation Object: Chia bức ảnh thành 3 vùng màu

```python
# Kmeans clustering with X
kmeans = KMeans(n_clusters=3).fit(X)
label = kmeans.predict(X)

# label of compressed image
compressed1D = np.zeros_like(X)

# Replace each pixel by its center
for k in range(3):
    compressed1D[label == k] = kmeans.cluster_centers_[k]
# Reshape 
compressed2D = compressed1D.reshape((img.shape[0], img.shape[1], img.shape[2]))
plt.imshow(compressed2D, interpolation = 'nearest')
plt.axis('off')
plt.show
```

Compression: Nén bức ảnh với số cluster lần lượt là 2, 5, 10, 20, 30, 50

```python
for K in [2,5,10,30,50,100]:
    print 'With', K, 'Clusters. We have the compressed image'
    # Kmeans clustering with X
    kmeans = KMeans(n_clusters=K).fit(X)
    label = kmeans.predict(X)
    # label of compressed image
    compressed1D = np.zeros_like(X)
    # Replace each pixel by its center
    for k in range(K):
        compressed1D[label == k] = kmeans.cluster_centers_[k]
    # Reshape 
    compressed2D = compressed1D.reshape((img.shape[0], img.shape[1], img.shape[2]))
    plt.imshow(compressed2D, interpolation = 'nearest')
    plt.axis('off')
    plt.show()
```
